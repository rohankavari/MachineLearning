```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.formula.api import ols
import statsmodels.api as sm
import pandas.util.testing as tm
import seaborn as sns
```

https://archive.ics.uci.edu/ml/datasets/Combined+Cycle+Power+Plant

```python
df=pd.read_excel("/content/drive/MyDrive/dataset/Folds5x2_pp.xlsx")
```

```python
df.describe()
```

```python
plt.figure(figsize=(6, 6))
sns.heatmap(df.corr(),annot=True)
```

```python
pp = sns.pairplot(data=df,y_vars=['PE'])
```

```python
mod = ols("PE~AT+V+AP+RH",data=df)
res = mod.fit()
print(res.summary())
```

```python
y_pred=res.fittedvalues
e=df['PE']-y_pred
```

```python
var='V'
plt.scatter(df[var], e ,c="green")
plt.plot([df[var].min(), df[var].max()], [0, 0], linewidth=2);
```

```python
var='AP'
plt.scatter(df[var], e ,c="green")
plt.plot([df[var].min(), df[var].max()], [0, 0], linewidth=2);
```
